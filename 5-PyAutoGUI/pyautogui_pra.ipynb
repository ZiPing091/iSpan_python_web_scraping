```python
import pyautogui
import time

"""
取得起始點的方式
1. time.sleep(2) 搭配 pyautogui.position() 取得滑鼠位置
2. 在 cmd 使用 python -m pyautogui 進入互動模式
"""

# 設定起始點
start_point = (522, 541)
# 移動滑鼠到起始點
pyautogui.moveTo(start_point[0], start_point[1])


# 按下滑鼠左鍵
distance = 200
while distance > 0:

    # 向右拖曳
    pyautogui.dragRel(distance, 0)
    distance -= 10
    # 向下拖曳
    pyautogui.dragRel(0, distance)
    
    # 向左拖曳
    pyautogui.dragRel(-distance, 0)
    distance -= 10
    # 向上拖曳
    pyautogui.dragRel(0, -distance)

# 放開滑鼠左鍵
```
