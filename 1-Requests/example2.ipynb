```python
import requests
from bs4 import BeautifulSoup as bs
import json
import os
# url = f"https://store.line.me/stickershop/product/32028/zh-Hant"
# r = requests.get(url)
# soup = bs(r.text, 'lxml')
```

```python

# create folder
folder_name = 'line_stickers'
if not os.path.exists(folder_name):
    os.mkdir(folder_name)
```

```python
'''
爬line貼圖
'''
def get_sticker(sticker_url):
    # url = f"https://store.line.me/stickershop/product/32028/zh-Hant"
    r = requests.get(sticker_url)
    soup = bs(r.text, 'lxml')

    sticker_name = soup.find('p', class_="mdCMN38Item01Ttl").get_text()


    sticker_folder = os.path.join(folder_name, sticker_name)
    if not os.path.exists(sticker_folder):
        os.mkdir(sticker_folder)


    sticker_list = soup.find('ul', class_="mdCMN09Ul FnStickerList")
    sticker_elements = sticker_list.find_all('li')
    # print(len(sticker_elements))

    for index, sticker_element in enumerate(sticker_elements, start = 1):
        print(sticker_element)
        # print(sticker_element['data-preview'])
        data_preview = json.loads(sticker_element['data-preview'])
        static_url = data_preview['staticUrl']

        image_content = requests.get(static_url).content
        with open(os.path.join(sticker_folder, f"{sticker_name}_{index}.png"), "wb") as f:
            f.write(image_content)

        # break
    print('done')
```

```python
get_sticker('https://store.line.me/stickershop/product/28048/zh-Hant')
```
